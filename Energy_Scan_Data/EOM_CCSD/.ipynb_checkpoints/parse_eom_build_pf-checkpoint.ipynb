```python
import h5py
import numpy as np
```

```python
def generate_lambda_vec_from_theta_and_phi(theta, phi):
    """ takes theta and phi in degrees and converts to radians 
    """
    # convert to radians
    theta *= np.pi / 180
    phi *= np.pi / 180
    
    # compute x, y, z
    x = np.sin(theta) * np.cos(phi)
    y = np.sin(theta) * np.sin(phi)
    z = np.cos(theta)

    return np.array([x, y, z])
```

```python
import h5py
import numpy as np

def parse_cq_h5_data(h5_file_path):
    print(f"Parsing EOMCC data from: {h5_file_path}\n")
    
    with h5py.File(h5_file_path, 'r') as f:
        # 1. Parse Scalars
        reference_energy = f['CC/REFERENCE_ENERGY'][()]
        correlation_energy = f['CC/CORRELATION_ENERGY'][()]
        
        # 2. Parse Excitation Energies Vector
        excitation_energies = f['CC/EXCITATION_ENERGIES'][:]
        num_excited_roots = len(excitation_energies)
        total_states = num_excited_roots + 1  # Including Ground State (Index 0)
        
        print("--- Scalar & Vector Quantities ---")
        print(f"Reference Energy:   {reference_energy}")
        print(f"Correlation Energy: {correlation_energy}")
        print(f"Excitation Energies Shape: {excitation_energies.shape} ({num_excited_roots} roots)\n")
        
        # 3. Read and Transpose Dipole Datasets 
        # (Transposing .T flips the row-major H5 layout to CQ's intended column-major layout)
        gs_dipole = f['CC/GROUND_STATE_DIPOLE'][:].T
        g2e_dipole = f['CC/GROUND_TO_EXCITED_TRANSITION_DIPOLE'][:].T
        e2g_dipole = f['CC/EXCITED_TO_GROUND_TRANSITION_DIPOLE'][:].T
        e2e_dipole = f['CC/EXCITED_TO_EXCITED_TRANSITION_DIPOLE'][:].T
        
        # --- Cartesian Axis Check ---
        # We need the 3 Cartesian components (X, Y, Z) to be the LAST dimension.
        # If the H5 layout had 3 as the first dimension, the .T operation made it the last.
        # Let's ensure the final dimension is indeed 3.
        def ensure_cartesian_last(arr, name):
            if arr.shape[0] == 3 and arr.shape[-1] != 3:
                # If 3 is at the front, move it to the back: (3, N, M) -> (N, M, 3)
                return np.moveaxis(arr, 0, -1)
            return arr

        gs_dipole = ensure_cartesian_last(gs_dipole, "Ground State")
        g2e_dipole = ensure_cartesian_last(g2e_dipole, "Ground to Excited")
        e2g_dipole = ensure_cartesian_last(e2g_dipole, "Excited to Ground")
        e2e_dipole = ensure_cartesian_last(e2e_dipole, "Excited to Excited")

        # 4. Initialize the Master Dipole Tensor (Total States, Total States, 3)
        master_dipole = np.zeros((total_states, total_states, 3))
        
        # Element (0, 0, :): Ground State Dipole
        # gs_dipole might be a 1D array of shape (3,) or 2D (1, 3). We flatten the state dimensions.
        master_dipole[0, 0, :] = gs_dipole.ravel()
        
        # Elements (0, 1:, :): Ground to Excited
        # CQ physics notation: < row | r | col > -> < 0 | r | excited >
        # Should shape-match (num_excited_roots, 3)
        master_dipole[0, 1:, :] = g2e_dipole.reshape(num_excited_roots, 3)
        
        # Elements (1:, 0, :): Excited to Ground
        # CQ physics notation: < row | r | col > -> < excited | r | 0 >
        # Should shape-match (num_excited_roots, 3)
        master_dipole[1:, 0, :] = e2g_dipole.reshape(num_excited_roots, 3)
        
        # Elements (1:, 1:, :): Excited to Excited
        # Should shape-match (num_excited_roots, num_excited_roots, 3)
        master_dipole[1:, 1:, :] = e2e_dipole.reshape(num_excited_roots, num_excited_roots, 3)
        
        print("--- Master Dipole Array Construction ---")
        print(f"Master Dipole Tensor Shape: {master_dipole.shape}")
        print("Successfully mapped all states!")
        
        return reference_energy, correlation_energy, excitation_energies, master_dipole

# --- Execution Block ---
if __name__ == "__main__":
    file_path = "nitrobenzene_eomrccsd.h5"  # <-- Change this to your file name
    
    try:
        ref_E, corr_E, excit_E, dipoles = parse_cq_h5_data(file_path)
        
        # Quick verification of a slice:
        print("\n--- Quick Sanity Check ---")
        print(f"Ground state dipole (X, Y, Z): {dipoles[0, 0, :]}")
        print(f"Transition dipole from GS to Excited State 1 (X, Y, Z): {dipoles[0, 1, :]}")
        
    except FileNotFoundError:
        print(f"Could not find the file '{file_path}'. Please verify the path.")
```

```python
# Swap only axis 0 (bra) and axis 1 (ket)
dipoles_transpose = np.transpose(dipoles, axes=(1, 0, 2))

# Create the symmetric transition dipole matrix
dipoles_sym = 0.5 * (dipoles + dipoles_transpose)

```

```python
def build_ladder_operator(dim):
    lower_mat = np.zeros((dim, dim))

    for i in range(dim-1):
        lower_mat[i, i+1] = np.sqrt(i+1)

    raise_mat = lower_mat.conj().T

    return lower_mat, raise_mat

```

```python
def build_PF_Hamiltonian(dim, omega, lambda_vec, e_el, dipoles):

    # dim_el -> number of electronic states, which is the first dimension of the dipole array
    dim_el = 10
    # build ladder operators
    a, a_dag = build_ladder_operator(dim)

    # build identities
    I_ph = np.eye(dim)
    I_el = np.eye(dim_el)

    # build bare electronic Hamiltonian
    H_el = np.diag(e_el[:dim_el])  # Ensure we only take as many energies as electronic states

    # build photon Hamiltonian
    H_ph = omega * (a_dag @ a + 0.5 * np.eye(dim))

    # build d matrix as lambda_vec dotted into dipoles 
    # where d[p,q] = \sum_i lambda_vec[i] * dipoles[p,q,i]
    d = np.einsum('i,pqi->pq', lambda_vec, dipoles)

    # build <d> matrix as d[0,0] * I_el
    d_0 = d[0, 0] * I_el

    # offset d by d_0
    d_cs = d[:dim_el, :dim_el] - d_0

    # build interaction Hamiltonian
    H_blc = -np.sqrt(omega / 2) * np.kron(d_cs, (a + a_dag))

    # H_dse 
    H_dse = 0.5 * np.kron(d_cs @ d_cs, I_ph)

    # total Hamiltonian
    H_total = np.kron(H_el, I_ph) + np.kron(I_el, H_ph) + H_blc + H_dse

    return H_total, np.kron(H_el, I_ph), np.kron(I_el, H_ph), H_blc, H_dse
```

```python
# empty array for electronic energies
dim_el = len(excit_E) + 1
E_el = np.zeros(dim_el)

# get ground state energy
E_el[0] = ref_E[0] + corr_E[0]

# get additional energies
for i in range(1, dim_el):
    E_el[i] = ref_E[0] + corr_E[0] + excit_E[i-1]




```

```python
import numpy as np
import pandas as pd

# Define your angles
theta_list = np.linspace(0, 180, 90)
phi_list = np.linspace(0, 360, 90)

# Container for our data rows
results = []
E_CCSD_GS = 0
print("Running Hamiltonian scan over angles...")
for theta in theta_list:
    for phi in phi_list:
        # 1. Calculate field coupling vector
        lambda_vec = generate_lambda_vec_from_theta_and_phi(theta, phi) * 0.1 
        
        # 2. Build and diagonalize Hamiltonian
        H_total, H_el, H_ph, H_blc, H_dse = build_PF_Hamiltonian(
            dim=6, omega=0.066148, lambda_vec=lambda_vec, e_el=E_el, dipoles=dipoles_sym
        )
        eigs, C = np.linalg.eigh(H_total)
        
        # 3. Compute expectation values for the polaritonic ground state (index 0)
        E_el_exp  = np.real(C[:, 0].conj().T @ H_el @ C[:, 0]) # Electronic
        E_ph_exp  = np.real(C[:, 0].conj().T @ H_ph @ C[:, 0])    # Photon
        E_blc_exp = np.real(C[:, 0].conj().T @ H_blc @ C[:, 0])   # Bilinear Coupling
        E_dse_exp = np.real(C[:, 0].conj().T @ H_dse @ C[:, 0])   # Dipole Self-Energy
        E_total   = eigs[0] # Total Ground State Energy
        E_CCSD_GS = H_el[0, 0]
        # Append data to list
        results.append({
            'theta': theta,
            'phi': phi,
            'E_el': E_el_exp,
            'E_el_uncoupled': E_CCSD_GS,
            'E_ph': E_ph_exp,
            'E_blc': E_blc_exp,
            'E_dse': E_dse_exp,
            'E_total': E_total
        })

# Save to CSV
df = pd.DataFrame(results)
df.to_csv("polariton_energy_scan.csv", index=False)
print("Data collection complete! Saved to 'polariton_energy_scan.csv'.")
```
